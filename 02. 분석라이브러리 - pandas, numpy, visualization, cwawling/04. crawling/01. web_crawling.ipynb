```python
# ì—¬ëŸ¬ ê°œì˜ ìš”ì†Œ ì¶”ì¶œ

html = """
<html><body>
  <ul>
    <li><a href="http://www.naver.com">naver</a></li>
    <li><a href="http://www.daum.net">daum</a></li>
  </ul>
</body></html>
"""

# HTML ë¶„ì„í•˜ê¸°
soup = BeautifulSoup(html)

# find_all() ë©”ì„œë“œë¡œ ì¶”ì¶œí•˜ê¸°
links = soup.find_all('a')

# ë§í¬ ëª©ë¡ ì¶”ì¶œí•˜ê¸°
for a in links:
    href = a.attrs['href']
    text = a.string
    print(text, '>', href)
```
